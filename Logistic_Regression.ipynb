```python
import pandas as pd
import numpy as np

data = pd.read_csv("/content/Iris.csv")

# Convert categorical target to numerical (0 for setosa, 1 for others)
data["species_encoded"] = (data["Species"] == "Iris-setosa").astype(int)

# Select features and target
X = data[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
y = data["species_encoded"]

# Add bias term (column of ones)
X = np.c_[np.ones((len(X), 1)), X]

# Define sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Define cost function (log loss)
def cost_function(h, y):
    return -np.mean(y * np.log(h) + (1 - y) * np.log(1 - h))

# Define gradient descent
def gradient_descent(X, y, learning_rate, num_iters):
    weights = np.zeros(X.shape[1])
    for i in range(num_iters):
        z = np.dot(X, weights)
        h = sigmoid(z)
        error = h - y
        gradient = np.dot(X.T, error) / len(y)
        weights -= learning_rate * gradient
    return weights

# Train the model
weights = gradient_descent(X, y, learning_rate=0.01, num_iters=1000)

# Make predictions
z = np.dot(X, weights)
predicted_probs = sigmoid(z)
predicted_class = np.round(predicted_probs).astype(int)


# Evaluate accuracy
accuracy = np.mean(predicted_class == y)
print("Accuracy:", accuracy)

# Predict on new data (modify X_new accordingly)
X_new = np.array([[5.1, 3.5, 1.4, 0.2]])
# Add bias term to new data
X_new = np.c_[np.ones((len(X_new), 1)), X_new]

# Predict on new data
z_new = np.dot(X_new, weights)
predicted_prob_new = sigmoid(z_new)
predicted_class_new = np.round(predicted_prob_new).astype(int)
print("Predicted class for new data:", predicted_class_new)

```
